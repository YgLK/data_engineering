# Cleaning data

```python
gdf = gdf.reset_index()
gdf
```

### The dataset is divided to nodes and ways subsets

```python
nodes = gdf.loc[gdf['element_type'] =='node']
nodes
```

```python
ways = gdf.loc[gdf['element_type'] =='way']
ways
```

### Dropping unnecessary column

```python
ways.drop(columns=['osmid'], inplace = True)
```

```python
ways
```

### Set default speed limit

```python
ways['maxspeed'].replace({'nan':'25 mph'}, inplace=True)
ways
```

### Change data type of speedlimit to int

```python
ways['maxspeed'] = ways.maxspeed.str.extract('(\d+)')
ways['maxspeed'] = ways['maxspeed'].astype(int)
```

```python
ways
```

### One hot encoding for oneway column

```python
dummies = pd.get_dummies(ways['oneway'],prefix="oneway")
```

```python
ways.drop(columns=['oneway'], inplace = True)
```

```python
ways = ways.join(dummies)
```

```python
ways
```

# Second attempt using geopandas.buffer

```python
gdfSpeed["buffered_geometry"] = gdfSpeed.buffer(0.0001)
```

```python
print("Type:", type(gdfSpeed))
print("Shape:", gdfSpeed.shape)
gdfSpeed.head()
```

```python
print("Type:", type(ways))
print("Shape:", ways.shape)
ways.head()
```

```python
buffered_result = gpd.sjoin(gdfSpeed.set_geometry("buffered_geometry"), ways[["name", "geometry"]], how='inner', predicate='contains')
```

```python
len(buffered_result)
```

### Drop duplicates base on geometry

```python
unique_result = buffered_result.reset_index()
unique_result.drop_duplicates(subset=["geometry", "buffered_geometry"], inplace=True)
unique_result
```

```python
fig, ax = plt.subplots(figsize=(18,20))

# Plot the footprint
unique_result.geometry.plot(ax=ax, facecolor='blue')
unique_result.buffered_geometry.plot(ax=ax, color='red', alpha=0.5)
plt.tight_layout()
```
